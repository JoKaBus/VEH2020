```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots()

ax.set_aspect(1)
ax.axis('off')

X = np.linspace(0,2*np.pi,1000)
ax.plot(np.sin(X),np.cos(X))

pre_cp = (-2+1j)
cp = pre_cp/np.sqrt(5)
ax.plot([0,cp.real,cp.real,0],[0,0,cp.imag,0])
ax.plot([0,pre_cp.real,pre_cp.real,0],[0,0,pre_cp.imag,0], '-.')

# plt.savefig('Plots/PresPlots/U43A1a.png')
```
