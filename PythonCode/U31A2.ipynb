```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 250
```

```python
def d(r): # plot discriminant
    return r**2*(r+3)**2

def x(x,r): # plot x\mapsto arctan(x^4+(3r-r^2)x^2 - 3r^3 for fixed r
    return np.arctan(x**4+(3*r-r**2)*x**2 - 3*(r**3))

def w(x,r): # plot w\mapsto arctan(w^2+(3r-r^2)w - 3r^3 for fixed r
    return np.arctan(x**2+(3*r-r**2)*x - 3*(r**3))

def w_1(r): # plot (r^2-3r)/2+r(r+3)
#     return np.arctan((r**2-3*r+r*(r+3))/2)
    return (r**2-3*r+r*(r+3))/2
    
def w_2(r): #  plot (r^2-3r)/2-r(r+3)
#     return np.arctan((r**2-3*r-r*(r+3))/2)
    return ((r**2-3*r)-r*(r+3))/2
```

```python
# plot discriminant D
fig, ax = plt.subplots()

X = np.linspace(-4,1,1000)
ax.set_xlim(-4,1)
ax.set_ylim(-1,7)

ax.plot([-4,1],[0,0]) # X-Axis
ax.plot([0,0],[-1,7]) # Y-Axis
ax.plot(X,d(X), label = '$x\mapsto r^2(r+3)^2$')
ax.legend(loc = 'upper center')

# plt.savefig('Plots/U31A2.png', dpi = 250) #  Sie mssen selbst einen Unterordner namens "Plots" anlegen
```

```python
# plot functions x,w
fig, ax = plt.subplots()

r = 1

plot_min = -10
plot_max = 10

X = np.linspace(plot_min,plot_max,1000)
# ax.set_ylim(-1,2.2)

ax.plot(X,x(X,r), label = '$x\mapsto$ arctan$(x^4+(3r-r^2)x^2 - 3r^3)$ for r = {}'.format(r))
ax.plot(X,w(X,r), label = '$w\mapsto$ arctan$(w^2+(3r-r^2)w - 3r^3)$ for r = {}'.format(r))

ax.legend()

# plt.savefig('Plots/U31A2r{}.png'.format(r), dpi = 250) # Sie mssen selbst einen Unterordner namens "Plots" anlegen
```

```python
# plot solutions
fig, ax = plt.subplots()

plot_min = -3
plot_max = 3

X = np.linspace(plot_min,plot_max,1000)
# ax.set_ylim(-1,2.2)

ax.plot(X,w_1(X), label = '$r\mapsto r^2$')
ax.plot(X,w_2(X), label = '$r\mapsto -3r$')
ax.plot([plot_min,plot_max],[0,0]) # X-Axis
ax.legend()

# plt.savefig('Plots/U31A2solution.png'.format(r), dpi = 250) # Sie mssen selbst einen Unterordner namens "Plots" anlegen
```
