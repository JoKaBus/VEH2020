```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots()

z1 = 2+5j; z2 = 2/5-7j

z = z1
z_abs = f'{np.abs(z):.3f}'; z_arg = f'{np.angle(z)/np.pi:.3f}'
ax.scatter(z.real,z.imag,label ='$|z_1| ≈ $'+z_abs+' $\phi$ ≈ '+z_arg+'$\pi$')
ax.plot([0,z.real],[0,z.imag])

z = z2
z_abs = f'{np.abs(z):.3f}'; z_arg = f'{np.angle(z)/np.pi:.3f}'
ax.scatter(z.real,z.imag,label ='$|z_2| ≈ $'+z_abs+' $\phi$ ≈ '+z_arg+'$\pi$')
ax.plot([0,z.real],[0,z.imag])
ax.legend()

z = z1+z2
z_abs = f'{np.abs(z):.3f}'; z_arg = f'{np.angle(z)/np.pi:.3f}'
ax.scatter(z.real,z.imag,label ='$|z_1+z_2| ≈ $'+z_abs+' $\phi$ ≈ '+z_arg+'$\pi$')
ax.plot([0,z.real],[0,z.imag])
ax.legend()

# z = z1/z2
# z_abs = f'{np.abs(z):.3f}'; z_arg = f'{np.angle(z)/np.pi:.3f}'
# ax.scatter(z.real,z.imag,label ='$|z_1/z_2| ≈ $'+z_abs+' $\phi$ ≈ '+z_arg+'$\pi$')
# ax.plot([0,z.real],[0,z.imag])
# ax.legend()

# plt.savefig('Plots/PresPlots/U51A1aplus.png')
```
