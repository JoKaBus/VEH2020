```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots()

z1 = 1-1j
z2 = ((1+1j)**13)/((1-1j)**7)
z3 = 1j**2014
z4 = (1-1j)**99

z = z3
ax.set_title('Aufgabenteil c)')
z_abs = f'{np.abs(z):.3f}'; 
z_arg = f'{np.angle(z)/np.pi:.3f}'

ax.scatter(z.real,z.imag,label ='$|z| ≈ $'+z_abs+' $\phi$ ≈ '+z_arg+'$\pi$')
ax.plot([0,z.real],[0,z.imag])

ax.scatter(z.real,0,label ='Re$z ≈ $'+z_abs)
ax.plot([0,z.real],[0,0])
ax.scatter(0,z.imag,label ='Im$z ≈ $'+z_abs)
ax.plot([0,0],[0,z.imag])

ax.plot([0,z.real],[0,-z.imag])
ax.scatter(z.real,-z.imag,label ='$\overline{z}$')
ax.legend()
```
