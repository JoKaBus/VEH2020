```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
def left_side(x):
    return np.arctan((2*x+1)/(x-2))

def right_side(x):
    return np.arctan((x+4)/(2*x+5))
```

```python
plot_min = -10
sig1 = -5/2
sig2 = 2
plot_max = 10
epsilon = 0.01

A = np.linspace(plot_min,sig1-epsilon,500)
B = np.linspace(sig1+epsilon,sig2-epsilon,500)
C = np.linspace(sig2+epsilon,plot_max,500)

fig, ax = plt.subplots()

ax.plot(A,left_side(A), color = 'blue', label = '$x\mapsto arctan((2x+1)/(x-2))$')
ax.plot(A,right_side(A), color = 'green', label = '$x\mapsto arctan((x+4)/(2x+5))$')
ax.plot(B,left_side(B), color = 'blue')
ax.plot(B,right_side(B), color = 'green')
ax.plot(C,left_side(C), color = 'blue')
ax.plot(C,right_side(C), color = 'green')
plt.legend()

plt.savefig('Plots/PresPlots/U32A2ungl.png')
```
