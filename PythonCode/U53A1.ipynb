```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
fig, ax = plt.subplots()

plt.title('bung 5.1 Aufgabe 1a')
ax.set_xlim(-0.1,2.1)
ax.set_ylim(-0.1,1.1)
ax.axis('off')

z = np.sqrt(3)+1j 
circ = 2*np.exp(1j*np.pi*np.linspace(0,1/12,100))
circ2 = 2*np.exp(1j*np.pi*np.linspace(1/12,2/12,100))

ax.plot([0,z.real],[0,0])
ax.plot([z.real,z.real],[0,z.imag])
ax.plot([0,z.real],[0,z.imag])
ax.plot(circ.real,circ.imag, '--')
ax.plot(circ2.real,circ2.imag, '--')

# plt.savefig('Plots/PresPlots/U53A1aempty.png')
```

```python
fig, ax = plt.subplots()

plt.title('bung 5.1 Aufgabe 1b')
ax.set_xlim(-17,4)
ax.set_ylim(-1,16)
ax.axis('off')

z = np.sqrt(3)+1j 

def my_circ(start,end,faktor):
    return faktor*np.exp(1j*np.pi*np.linspace(start,end,100))

ax.scatter([-16],[0])

for k in range(4):
    circ = my_circ(k/4,(k+1)/4,(k+1)**2)
    ax.plot(circ.real,circ.imag, '--')
    ax.plot([-(k),-((k+1)**2)],[0,0])
#     plt.savefig('Plots/PresPlots/U53A1bcirc1{}.png'.format(k))
    
# plt.savefig('Plots/PresPlots/U53A1bempty.png')
```
