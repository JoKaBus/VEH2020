```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 300
```

```python
def parabola(x,a,x0,y0):
    return a*(x-x0)**2+y0

def y1():
    a = 1; x0 = 3; y0 = 1
    return a, x0, y0

def y2():
    a = 1/4; x0 = 2; y0 = 3
    return a, x0, y0

def y22(x):
    return (1/4)*x**2-x+4

def y33(x):
    return -(2/3)*(x**2)+x

def y3():
    a = -2/3; x0 = 3/4; y0 = (2/3)*(3/4)**2
    return a, x0, y0
```

```python
a, x0, y0 = y1()

plot_min = -1; plot_max = 4
X = np.linspace(plot_min,plot_max,1000)

plt.title('$x\mapsto a(x-x_0)^2+y_0$')
plt.plot(X,parabola(X,a,x0,y0), label = 'Let $a = {}, x_0 = {}, y_0 = {}$'.format(a,x0,y0))
plt.legend()

plt.savefig('Plots/PresPlots/U42A2a.png')
```

```python
a, x0, y0 = y2()

X = np.linspace(plot_min,plot_max,1000)

plt.title('$x\mapsto a(x-x_0)^2+y_0$')
plt.plot(X,parabola(X,a,x0,y0), label = 'Let $a = {}, x_0 = {}, y_0 = {}$'.format(a,x0,y0))
plt.legend()

plt.savefig('Plots/PresPlots/U42A2b.png')
```

```python
a, x0, y0 = y3()

X = np.linspace(plot_min,plot_max,1000)

plt.title('$x\mapsto a(x-x_0)^2+y_0$')
plt.plot(X,parabola(X,a,x0,y0), label = 'Let $a = {}, x_0 = {}, y_0 = {}$'.format(a,x0,y0))
# plt.legend()

plt.savefig('Plots/PresPlots/U42A2c.png')
```
